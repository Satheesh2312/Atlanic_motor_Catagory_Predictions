```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_excel('Test_Data.xlsx')
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop('Category',axis=1,inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.dropna()
```

```python
df['Amount Month 1']=df['Amount Month 1'].fillna(df['Amount Month 1'].mean(),axis=0)
df['Amount Month 3']=df['Amount Month 3'].fillna(df['Amount Month 3'].mean(),axis=0)
```

```python
df.isnull().sum()
```

```python
def df_Encode(df):
    for column in df.columns[~df.columns.isin(['Amount Month 1','Amount Month 2','Amount Month 3','Amount Month 4'])]:
        df[column]=df[column].factorize()[0] ## Factorize method is used for numerical representation
                                                        ##of an array having categorical features
    return df
df_copy=df_Encode(df.copy())
```
